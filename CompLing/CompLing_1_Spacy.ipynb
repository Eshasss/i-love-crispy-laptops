```python
words = word_tokenize(text, language='russian')
```

```python
morph.parse('Иванов')[0].lexeme # Фамилия часто встречается, знает мн.ч.
```

```python
def proper_check(word: str, new_word: str): # Если слово было именем собственным(с большой буквы), после парсинга оно теряет заглавную букву. Функция это фиксит
    if word[0].isupper():
        return new_word.title()
    return new_word
```

```python
def solution(tokens: list):
    for i in range(len(tokens)):
        token = tokens[i]
        token_parsed = morph.parse(token)[0]
        if ('NOUN' in token_parsed.tag) or ('ADJF' in token_parsed.tag):
            try:
                tokens[i] = proper_check(token, token_parsed.inflect({'plur'})[0]) # Необходимо, тк для некоторых им. собственных pymorphy не умеет образовывать мн.ч.
            except:
                pass
        elif 'VERB' in token_parsed.tag:
            tokens[i] = proper_check(token, token_parsed.inflect({'plur', 'past'})[0])

    return tokens

```

```python
new_text = solution(words)
new_text = ' '.join(new_text)
new_text = re.sub(r'''\s([^\w\s|—"])''',r'\1', new_text) # отдельно случай с тире
new_text = re.sub(r"``\s(.*?)''", r' "\1"', new_text)
new_text = re.sub(r'''\(\s(.*?)\)''',r'(\1)', new_text) # Случаи с кавычками и скобочками, хотя в этом тексте их нет
```

```python
def paradigm(words: list): # Делает парадигмы для слов
    paradigms = []
    for w in words:
        for i in range(len(morph.parse(w))):
            paradigms.append(morph.parse(w)[0].lexeme[i].word)
    return paradigms


months = paradigm(["январь", "февраль", "март", "апрель", "май", "июнь", "июль", "август", "сентябрь", "октябрь", "декабрь"])
year = paradigm(["год"])
year.append("г.")
```

```python
ru_nlp = spacy.load("ru_core_news_sm")
ruler = ru_nlp.add_pipe("entity_ruler")

patterns = [
    {"label": "DATE", "pattern": [{"IS_DIGIT": True}, {"TEXT": "-"}, {"IS_DIGIT": True}, {"TEXT": "-"}, {"IS_DIGIT": True}]},  # дд-мм-гггг
    {"label": "DATE", "pattern": [{"IS_DIGIT": True}, {"TEXT": "/"}, {"IS_DIGIT": True}, {"TEXT": "/"}, {"IS_DIGIT": True}]},  # дд/мм/гггг
    {"label": "DATE", "pattern": [{"IS_DIGIT": True, "LENGTH": 4}, {"TEXT": {"IN": year}}]}, # 1234 г. 1234 год
    {"label": "DATE", "pattern": [{"IS_DIGIT": True, "LENGTH": 4}, {"TEXT": {"IN": ["—", "–", "-", "—"]}}, {"IS_DIGIT": True, "LENGTH": 4}]}, # гггг-гггг
    {"label": "DATE", "pattern": [{"IS_DIGIT": True}, {"TEXT": {"IN": months}}]}  # 8 ноября
]

ruler.add_patterns(patterns)

doc = ru_nlp("Пушкин родился 26 мая 1799 г. в Москве, в 323 Немецкой слободе. 34 Квартира . 02-12-3021 Шесть лет (1811— 1817) Пушкин провёл в Императорском Царскосельском лицее, открытом 19 октября 1811 года.")

displacy.render(doc, style='ent', jupyter=True)
```

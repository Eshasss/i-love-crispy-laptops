```python
from ipaddress import *

net = ip_network(f'192.168.32.160/255.255.255.240',0) #создается так
# или так:
# net = ip_network(f'1.2.3.4/26', 0) #потому что в такой маска 26 единиц
print(net) # адрес сети
print(net[0]) # Адрес сети
print(net[-1]) #адрес широкого вещания сети
print(net.netmask) # маска сети
print(net.num_addresses) # Количество IP-адресов в сETU
print(net.num_addresses - 2) # количество узолой в сети
# получить все IP-адреса сети
s=0
for ip in net:
    k = int(f"{ip:b}")
    if k%2==0:
        s += 1

print(s)
```

```python
from ipaddress import *

net = ip_network(f'192.168.63.0/255.255.255.128',0)
s =0
for ip in net:
    k = f"{ip:b}"
    print(k)
    r= 0
    for t in k:
        if t == "0":
            r += 1
    if r%5 != 0:
        s+=1
print(s)

```

```python
from ipaddress import *

net = ip_network(f'192.168.0.0/255.255.192.0',0)
print(net) # адрес сети
print(net[0]) # Адрес сети
print(str(net[0]).split("."))
print(f"{net[0]:b}")
print(len(f"{net[0]:b}"))
s = 0
for ip in net:
    n = f"{ip:b}"
    if n.count("0") < n.count("1"):
        s += 1
print(s)
```

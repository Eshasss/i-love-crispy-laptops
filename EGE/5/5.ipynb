```python
def f(n): # в 3х
    s = ""
    while n != 0:
        s += str(n%3)
        n = n//3
    return s[::-1]
answs = []
for i in range(3,400):
    nn = f(i)
    ns = sorted(nn, key=int)[::-1]
    m = max(ns)
    ns += m
    k = ""
    for i in ns:
        k += i
    answs.append(int(k,3))
print(sorted(answs))
```

```python
import sys
sys.setrecursionlimit(10001)
from functools import lru_cache

@lru_cache(None)
def g(n):
    if n <= 25:
        return 2*(n+1)
    elif n > 25:
        return g(n-4) + n

@lru_cache(None)
def f(n):
    return g(n-1) - g(n-5)


print(f(150774))

```

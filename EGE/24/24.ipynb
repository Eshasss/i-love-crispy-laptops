```python
import re
sequence = open('EGE/333/24.txt').read().strip()
seqs = [seq for seq in re.findall(r'(?:[1-9][0-9]*|0)(?:[+*](?:[1-9][0-9]*|0))+', sequence)]
good = []
for seq in seqs:
    nums = 0
    znak = 0
    for s in seq:
        if s == "+" or s == "*":
            nums += 1
            znak += 1
    #print(nums, znak)
    if nums <= 50 and znak <= 49:
        good.append(seq)


lgood = [len(k) for k in good]
print(good)
print(lgood[0])
print(max(lgood))

```

```python
with open("EGE/242444.txt") as f:
    we = f.read().split("\n")
www = []

for w in we:
    if w.count("A") < 25:
        www.append(w)

def f(cc):
    alph = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    dists = []
    for a in alph:
        first = cc.find(a)
        ilast = first
        for c in range(len(cc)):
            if cc[c] == a:
                ilast = c
        last = ilast
        dists.append(last-first)
    #print(dists)
    return max(dists)
for k in www:
    print(f(k))


```

```python
from turtle import *
tracer(0)

k=15

left(90)

right(90)

for i in range(2025):

    forward(10*k)
    right(90)
    forward(5*k)
    right(90)
penup()
forward(7*k)
right(120)
pendown()
for i in range(6):
    forward(5*k)
    right(120)

penup()
for x in range(-30, 30):
    for y in range(-30, 30):
        setpos(x*k, y*k)
        dot(4, 'red')

done()
```

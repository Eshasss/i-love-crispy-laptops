```python
def f (n):
    nbin = bin(n)[2::]
    nbinnew = ""
    for k in nbin:
        if k=="1":
            nbinnew += "0"
        else:
            nbinnew += "1"
    nnewint = int(nbinnew, 2)
    return n - nnewint
```

```python
for i in range( 1, 100000):
    s  = f(i)
    if s== 999:
        print(s, i)
```
